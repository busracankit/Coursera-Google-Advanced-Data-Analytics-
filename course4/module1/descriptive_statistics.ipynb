```python
import pandas as pd
import numpy as np
```

```python
epa_data = pd.read_csv("c4_epa_air_quality.csv", index_col = 0)
```

```python
epa_data.head(10)
```

```python
epa_data.describe()
```

```python
# 'state_name' sütunu için kategorik özet bilgileri al (kaç tane, kaçı eşsiz, en sık görüleni ve tekrar sayısı)
epa_data["state_name"].describe()
```

```python
# 'aqi' sütununun (hava kalitesi indeksi) ortalama değerini hesapla
np.mean(epa_data["aqi"])
```

```python
# 'aqi' sütununun (hava kalitesi indeksi) medyan (ortanca) değerini hesapla
np.median(epa_data["aqi"])
```

```python
# 'aqi' sütunundaki (hava kalitesi indeksi) en düşük değeri bul
np.min(epa_data["aqi"])
```

```python
# 'aqi' sütunundaki (hava kalitesi indeksi) en yüksek değeri bul
np.max(epa_data["aqi"])
```

```python
# 'aqi' sütununun (hava kalitesi indeksi) standart sapmasını hesapla (örneklem için)
np.std(epa_data["aqi"], ddof=1)
```
