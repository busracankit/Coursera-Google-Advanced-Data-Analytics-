```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
# "Chinstrap" türüne ait penguenleri seçiyoruz ve indeksleri sıfırlıyoruz.
# Böylece seçilen verilerin indeksleri 0'dan başlayarak düzgün bir şekilde sıralanmış oluyor.

chinstrap_penguins = penguins[penguins["species"] =="Chinstrap"]

chinstrap_penguins.reset_index(inplace = True, drop=True)
```

```python
# Chinstrap penguenlerinden "bill_depth_mm" ve "flipper_length_mm" sütunlarını seçiyoruz.

ols_data = chinstrap_penguins[["bill_depth_mm", "flipper_length_mm"]]
```

```python
# Bağımlı değişkenin "flipper_length_mm", bağımsız değişkenin "bill_depth_mm" olduğu bir regresyon formülü tanımlıyoruz.

ols_formula = "flipper_length_mm ~ bill_depth_mm"
```

```python
from statsmodels.formula.api import ols
```

```python
# Tanımladığımız formül ve verilerle OLS (Ordinary Least Squares) regresyon modeli oluşturuyoruz.
# Modeli eğitiyoruz (fit) ve özet sonuçları (summary) görüntülüyoruz.

OLS = ols(formula = ols_formula, data = ols_data)
model = OLS.fit()
model.summary()
```

```python
# Eğittiğimiz modeli kullanarak "bill_depth_mm" değerlerine göre tahminler yapıyoruz.

predictions = model.predict(chinstrap_penguins[["bill_depth_mm"]])
```

```python
# Modelin tahmin hatalarını (residuals) elde ediyoruz.

residuals = model.resid
```
