```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
# Chinstrap türünü filtreledik, eksik (NaN) verileri sildik ve indeksleri sıfırdan başlattık

penguins_sub = penguins[penguins["species"] != "Chinstrap"]
penguins_final = penguins_sub.dropna()
penguins_final.reset_index(inplace=True, drop=True)
```

```python
# Sayısal sütunlar arasında çift değişkenli ilişkileri ve dağılımları pairplot ile göster

sns.pairplot(penguins_final)
```

```python
# Regresyon analizinde kullanılacak iki değişkeni seç

ols_data = penguins_final[["bill_length_mm", "body_mass_g"]]
```

```python
# Regresyon formülünü tanımla

ols_formula = "body_mass_g ~ bill_length_mm"
```

```python
# Regresyon modeli için ols fonksiyonunu yükle

from statsmodels.formula.api import ols
```

```python
# OLS yöntemiyle model oluşturuluyor ve veriye göre eğitiliyor

OLS = ols(formula = ols_formula, data = ols_data)
model = OLS.fit()
```

```python
# Model özeti (istatistiksel analiz raporu)
model.summary()
```

```python
# Bill length ve body mass arasındaki ilişkiyi regresyon çizgisi ile görselleştir

sns.regplot(x = "bill_length_mm", y = "body_mass_g", data = ols_data)
```

```python
# Gaga uzunluğuna göre tahmin edilen vücut kütlesi değerlerini hesapladık

X = ols_data["bill_length_mm"]

fitted_values = model.predict(X)
```

```python
# Modelin kalıntılarını (gerçek ve tahmin edilen değer farklarını) hesapla

residuals = model.resid
```

```python
import matplotlib.pyplot as plt
fig = sns.histplot(residuals)
fig.set_xlabel("Residual Value")
fig.set_title("Histogram of Residuals")
plt.show()
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
fig = sm.qqplot(model.resid, line = 's')
plt.show()
```

```python
# Fitted values ve residuals arasındaki ilişkiyi scatter plot ile görselleştirdik
# Modelin hatalarını analiz ettik

import matplotlib.pyplot as plt
fig = sns.scatterplot(x=fitted_values, y=residuals)

fig.axhline(0)

fig.set_xlabel("Fitted Values")
fig.set_ylabel("Residuals")

plt.show()
```
